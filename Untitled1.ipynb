<h3>FUNKCIJE</h3>

```python
def eval(y, y_predicted):
    print(f'Accuracy: {round(accuracy_score(y, y_predicted)*100,2)}%')
    print(f'Precision: {round(precision_score(y, y_predicted)*100,2)}%')
    print(f'Recall: {round(recall_score(y, y_predicted)*100,2)}%')
    print(f'F-score: {round(f1_score(y, y_predicted)*100,2)}%')
    print(f'Auc: {round(roc_auc_score(y, y_predicted)*100,2)}%')
```

```python
def matrica_konfuzije(y_true, y_predict):
    ax=plt.subplot()
    sns.heatmap(confusion_matrix(y_true, y_predict), ax=ax, annot=True, fmt='g')
    ax.set_xlabel('Predicted')
    ax.set_ylabel('True')
```

```python
def kros_model(model, X_za_tr, y_za_tr):
    X_za_tr.reset_index(drop=True, inplace=True)
    y_za_tr.reset_index(drop=True, inplace=True)

    krossValidation= KFold(n_splits=10, shuffle=True, random_state=2020)
    results=[]
    acc=[]
    prec=[]
    rec=[]
    f1=[]
    auc=[]
    for tr_index, ts_index in krossValidation.split(X_za_tr):
        X_tr, y_tr= X_za_tr.loc[tr_index,:], y_za_tr.loc[tr_index]
        X_ts, y_ts= X_za_tr.loc[ts_index,:], y_za_tr.loc[ts_index]
        model.fit(X_tr, y_tr)
        y_predict= model.predict(X_ts)
        
        matrica_konfuzije(y_ts, y_predict)
        
        acc.append(accuracy_score(y_ts, y_predict))
        prec.append(precision_score(y_ts, y_predict))
        rec.append(recall_score(y_ts, y_predict))
        f1.append(f1_score(y_ts, y_predict))
        auc.append(roc_auc_score(y_ts, y_predict))

        #results.append(eval(y_ts, y_predicted))
        #print('===============================')
    print(f'Accuracy iznosi {round(np.mean(acc)*100, 2)}% +/- {round(np.std(acc)*100,2)}%')
    print(f'Precision iznosi {round(np.mean(prec)*100, 2)}% +/- {round(np.std(prec)*100,2)}%')
    print(f'Recall iznosi {round(np.mean(rec)*100, 2)}% +/- {round(np.std(rec)*100,2)}%')
    print(f'F-score iznosi {round(np.mean(f1)*100, 2)}% +/- {round(np.std(f1)*100,2)}%')
    print(f'Auc iznosi {round(np.mean(auc)*100, 2)}% +/- {round(np.std(auc)*100,2)}%')
```

```python
def primena_na_testu(model, X_za_test, y_za_test):
    y_hat=model.predict(X_za_test)
    eval(y_za_test, y_hat)
    matrica_konfuzije(y_test, y_hat)
    
```

```python
X= insurance.iloc[:, :-1]
y=insurance['charges']
X.head()
y.head()
```

```python
X_tr, X_ts, y_tr, y_ts= train_test_split(X,y,test_size=0.3, random_state=2020)
```

```python
kategorijske_kol=list(X_tr.select_dtypes(include='object').columns)
numericke_kol=list(X_tr.select_dtypes(exclude='object').columns)
```

```python
num_preparation=SimpleImputer(strategy='median')

categ_preparation= Pipeline(steps=[('imputer', SimpleImputer(strategy='most_frequent')),
                                  ('one_hot_encoding', OneHotEncoder(handle_unknown='ignore'))])

pre_pipeline=ColumnTransformer(transformers=[('num', num_preparation, numericke_kol),
                        ('categ', categ_preparation, kategorijske_kol)])

dt = DecisionTreeRegressor()

pipeline= Pipeline(steps=[('preprocesing', pre_pipeline),
                         ('decision tree regressor', dt)])
```

```python
plt.hist(y_tr)
```

```python
y_tr_log=np.log(y_tr)
plt.hist(y_tr_log)
```

```python
def rmse_score(y_t, y_p):
    razlika= y_t-y_p
    kvadrat= np.power(razlika,2)
    sr_vr= kvadrat.mean()
    rmse= np.sqrt(sr_vr)
    return (rmse)

rmse = make_scorer(rmse_score, greater_is_better=False)

def unlog_rms(y_t, y_p):
    y_un=np.exp(y_p)
    rmse1=rmse_score(y_t, y_un)
    return rmse1

unlog_rmse= make_scorer(unlog_rms, greater_is_better=False)
```

```python
cross_val_lin=cross_validate(pipeline, X_tr, y_tr_log, cv=5,  scoring={ 'MAE':'neg_mean_absolute_error', 'MSE':'neg_mean_squared_error', 'R2':'r2','RMSE' : unlog_rmse})
print(cross_val_lin)
```

```python
def mere_lin(kv): 
    mere={}
    for mesure, performance in kv.items():
        mere.update({mesure:(np.mean(performance), np.std(performance))})
    return mere
mere_lin(cross_val_lin)
```

```python
cross_val_lin=cross_validate(pipeline, X_tr, y_tr_log, cv=5,  scoring={ 'MAE':'neg_mean_absolute_error', 'MSE':'neg_mean_squared_error', 'R2':'r2','RMSE' : unlog_rmse})
print(cross_val_lin)
```

<h4>Optimizovanje parametara</h4>

```python
max_depths=[2, 4, 6, 8]
min_samples_leafs=[10, 20, 30 ,40]

for md in max_depths:
    for msl in min_samples_leafs:
        opt_tree= DecisionTreeRegressor(max_depth=md, min_samples_leaf=msl)
        opt_pipeline= Pipeline(steps=[('preprocesing', pre_pipeline),
                         ('decision tree regressor', opt_tree)])
        cross_val_opt=cross_validate(opt_pipeline, X_tr, y_tr, cv=5,  scoring={ 'MAE':'neg_mean_absolute_error', 'MSE':'neg_mean_squared_error', 'R2':'r2','RMSE' : rmse})
        print(f"Max depth: {md}, Min samples leaf: {msl}, RMSE: {mere_lin(cross_val_opt).get('test_RMSE')}")
```
